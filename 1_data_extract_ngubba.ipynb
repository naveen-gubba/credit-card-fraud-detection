```python
#change dtype of columns in boolean string to bool
df_bool_columns = ['expirationDateKeyInMatch','isFraud','cardPresent']

for col in df_bool_columns:
    data_df.loc[:,col]= data_df.loc[:,col].replace({'false' : False, 'true' : True})
```

```python
#change dtype of columns in float string to float
df_float_col = ['creditLimit','availableMoney','transactionAmount','currentBalance']

#change dtype of columns in int string to int
df_int_col = ['cardCVV','enteredCVV','cardLast4Digits']

for col in df_float_col:
    data_df.loc[:,col]= data_df.loc[:,col].astype(float)

for col in df_int_col:
    data_df.loc[:,col]= data_df.loc[:,col].astype(int)
```

```python
#change dtype of columns in date string to date
df_date_col = ['transactionDateTime','currentExpDate','accountOpenDate','dateOfLastAddressChange']

for col in df_date_col:
    data_df.loc[:,col]= data_df.loc[:,col].astype('datetime64[ns]')
```
