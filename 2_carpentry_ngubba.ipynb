```python
pattern = r"(\w.*[^#0-9]+)"

unique_merchant = list(set(data.loc[:,'merchantName']))

unique_dict = {}

for i in range(len(unique_merchant)):
    unique_dict[unique_merchant[i]] = re.findall(pattern,unique_merchant[i])[0].strip()
    
merchant_name_uq = []

for i in range(data.shape[0]):
    merchant_name_uq.append(unique_dict[data.loc[i,'merchantName']])

data['merchant_name_uq'] = merchant_name_uq
```

```python
data.columns = [i.lower() for i in data.columns]
```

```python
data = data.loc[:,data.columns.sort_values()]
data.head().T
```

```python
data['posentrymode'].unique()
```

```python
def online_instore(x):
    if x == 2 or x == 5 or x ==90:
        return 'instore'
    
    elif x == 9 or x == 80:
        return 'electronic'
    else:
        return 'missing'

data['iselectronic'] = data['posentrymode'].apply(online_instore)
```
